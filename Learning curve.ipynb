```python
import keras 
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
#Initializing the ANN
classifier = Sequential()

#Adding input and 1st hidden layer with Dropout
classifier.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'tanh', input_dim = 1))
#classifier.add(Dropout(rate = 0.1))

# 2nd hidden layer
classifier.add(Dense(units = 16, kernel_initializer = 'uniform', activation = 'tanh'))
#classifier.add(Dropout(rate = 0.1))

#3r hiddern Layer
classifier.add(Dense(units = 32, kernel_initializer = 'uniform', activation = 'tanh'))

#output Layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'linear'))

#Compiling
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'] )
```

```python
from scipy import interpolate
```

```python
tck1 = interpolate.splrep(x[0:199], a1[0:199], s = 1000)
ynew1 = interpolate.splev(x[0:199], tck1, der = 0)
```

```python
plt.plot(ynew1, label='spline curve')
plt.plot(x, y, label='loss')
plt.legend()
```

```python
np.std(a1-ynew1)
```
