```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
```

# Single Learning Rate

```python
# Validation loss
val_loss = np.load('./data/XceptionNet_loss0-80.npy')

plt.figure(figsize=(10,5))
plt.grid()
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.plot(val_loss,color='red')
```

```python
# Validation loss
val_acc = np.load('./data/XceptionNet_acc0-80.npy')

plt.figure(figsize=(10,5))
plt.grid()
plt.xlabel('Epochs')
plt.ylabel('accuracy')
plt.plot(val_acc,color='red')
```

```python
# Max Accuracy reached
np.argmax(val_acc)
```
