```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
```

# Single Learning Rate

```python
# Validation Loss
val_loss = np.load('./data/VGG16Slr_loss0-80.npy')

plt.figure(figsize=(10,5))
plt.grid()
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.plot(val_loss,color='red')
```

```python
# Validation Accuracy
val_acc = np.load('./data/VGG16Slr_acc0-80.npy')

plt.figure(figsize=(10,5))
plt.grid()
plt.xlabel('Epochs')
plt.ylabel('accuracy')
plt.plot(val_acc,color='red')
```

```python
# Max Accuracy reached
print(val_acc.max())
np.argmax(val_acc)
```

# Learning Rate Decay

```python
# Validation loss
val_loss2 = np.load('./data/VGG16lrd_loss0-80.npy')

plt.figure(figsize=(10,5))
plt.grid()
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.plot(val_loss2,color='green')
```

```python
# Validation accuracy
val_acc2 = np.load('./data/VGG16lrd_acc0-80.npy')

plt.figure(figsize=(10,5))
plt.grid()
plt.xlabel('Epochs')
plt.ylabel('accuracy')
plt.plot(val_acc2,color='green')
```

```python
# Max Accuracy reached
print(val_acc2.max())
np.argmax(val_acc2)
```

# Step Reduced Learning Rate

```python
# Validation loss
val_loss3 = np.load('./data/VGG16Srlr_loss0-100.npy')

plt.figure(figsize=(10,5))
plt.grid()
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.plot(val_loss3)
```

```python
# Validation loss
val_acc3 = np.load('./data/VGG16Srlr_acc0-100.npy')

plt.figure(figsize=(10,5))
plt.grid()
plt.xlabel('Epochs')
plt.ylabel('accuracy')
plt.plot(val_acc3)
```

# Comparing the methods

```python
plt.figure(figsize=(15,7.5))
plt.grid()
plt.xlabel('Epochs')
plt.ylabel('accuracy')
plt.plot(val_acc3, label='SrLr')
plt.plot(val_acc2, color='green', label='LrD')
plt.plot(val_acc, color='red', label='SLr')
plt.legend(loc="lower right")
```

```python
plt.figure(figsize=(10,12))
plt.grid()
plt.xlabel('Epochs')
plt.ylabel('accuracy')
plt.plot(val_acc3+1, label='SrLr')
plt.plot(val_acc2+.5, color='green', label='LrD')
plt.plot(val_acc, color='red', label='SLr')
plt.legend(loc="lower right")
```
